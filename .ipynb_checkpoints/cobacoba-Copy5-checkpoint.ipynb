```python
# =============================================================================
# KODE PREDIKSI UNTUK DATA BERSIH (TANPA PREPROCESSING LAGI)
# =============================================================================

import pandas as pd
import numpy as np
import joblib
import matplotlib.pyplot as plt

# =============================================================================
# 1. LOAD MODEL DAN ARTIFACTS YANG SUDAH DITRAINING
# =============================================================================
print("1. MEMUAT MODEL DAN ARTIFACTS...")
print("=" * 50)

try:
    # Load semua artifacts yang sudah disimpan
    model = joblib.load('final_model_astacita_clean.pkl')
    vectorizer = joblib.load('tfidf_vectorizer_astacita_clean.pkl')
    
    print("âœ… Model dan vectorizer berhasil dimuat")
    print(f"   Model: {type(model).__name__}")
    print(f"   Jumlah fitur: {len(vectorizer.get_feature_names_out())}")
    
except FileNotFoundError as e:
    print(f"âŒ File tidak ditemukan: {e}")
    print("Pastikan file berikut ada di directory:")
    print("- final_model_astacita_clean.pkl")
    print("- tfidf_vectorizer_astacita_clean.pkl")
    exit()

# =============================================================================
# 2. LOAD DATA TEST YANG SUDAH DIBERSIHKAN
# =============================================================================
print("\n2. MEMUAT DATA TEST BERSIH...")
print("=" * 50)

try:
    # Asumsi data sudah dibersihkan dan siap diproses
    test_data = pd.read_csv('test_clean.csv')  # Ganti dengan file data bersih Anda
    print(f"âœ… Data test bersih berhasil dimuat: {test_data.shape}")
    print(f"   Columns: {list(test_data.columns)}")
    print(f"   Sample data:")
    print(test_data.head())
    
except FileNotFoundError:
    print("âŒ File test_cleaned.csv tidak ditemukan")
    print("Pastikan data sudah dibersihkan dan disimpan sebagai test_cleaned.csv")
    exit()

# Cek kolom yang tersedia dan cari kolom teks yang sesuai
text_columns = [col for col in test_data.columns if 'teks' in col.lower() or 'text' in col.lower() or 'clean' in col.lower()]

if not text_columns:
    print("âŒ Tidak ditemukan kolom teks yang sesuai")
    print("Kolom yang tersedia:", list(test_data.columns))
    print("Silakan rename kolom teks bersih menjadi 'teks_bersih' atau edit kode untuk menggunakan kolom yang sesuai")
    exit()

# Gunakan kolom teks pertama yang ditemukan
text_column = text_columns[0]
print(f"âœ… Menggunakan kolom '{text_column}' sebagai teks input")

if 'id' not in test_data.columns:
    print("âš   Kolom 'id' tidak ditemukan, akan menggunakan index sebagai id")
    test_data['id'] = test_data.index

# =============================================================================
# 3. TRANSFORM TEKS BERSIH MENJADI FITUR TF-IDF
# =============================================================================
print(f"\n3. TRANSFORMASI KOLOM '{text_column}' KE FITUR TF-IDF...")
print("=" * 50)

# Periksa apakah ada nilai NaN dalam kolom teks
if test_data[text_column].isna().any():
    print("âš   Ditemukan nilai NaN dalam kolom teks, mengisi dengan string kosong")
    test_data[text_column] = test_data[text_column].fillna('')

# Gunakan vectorizer yang SAMA dengan training (jangan fit ulang!)
X_test_tfidf = vectorizer.transform(test_data[text_column])

print(f"âœ… Transformasi selesai. Shape: {X_test_tfidf.shape}")
print(f"   Sparsity: {100 * (1 - X_test_tfidf.nnz / (X_test_tfidf.shape[0] * X_test_tfidf.shape[1])):.2f}%")

# =============================================================================
# 4. PREDIKSI LABEL DENGAN MODEL YANG SUDAH DILATIH
# =============================================================================
print("\n4. MELAKUKAN PREDIKSI LABEL...")
print("=" * 50)

# Prediksi label
predictions = model.predict(X_test_tfidf)

# Prediksi probabilitas (tingkat keyakinan)
if hasattr(model, 'predict_proba'):
    probabilities = model.predict_proba(X_test_tfidf)
    confidence_scores = np.max(probabilities, axis=1)
else:
    confidence_scores = np.ones(len(predictions))  # Default confidence 1.0

# Tambahkan hasil prediksi ke DataFrame
test_data['predicted_label'] = predictions
test_data['confidence'] = confidence_scores

print("âœ… Prediksi selesai!")
print(f"   Jumlah sample yang diprediksi: {len(test_data)}")
print("\nDistribusi prediksi:")
print(test_data['predicted_label'].value_counts().sort_index())

# =============================================================================
# 5. TAMPILKAN HASIL PREDIKSI
# =============================================================================
print("\n5. HASIL PREDIKSI:")
print("=" * 50)

# Tampilkan beberapa contoh prediksi
print("Contoh hasil prediksi:")
print("-" * 80)
for i in range(min(5, len(test_data))):
    row = test_data.iloc[i]
    print(f"ID: {row['id']}")
    print(f"Teks: {str(row[text_column])[:70]}...")
    print(f"Label Prediksi: {row['predicted_label']}")
    print(f"Confidence: {row['confidence']:.3f}")
    print("-" * 80)

# Statistik confidence
print(f"\nðŸ“Š Statistik Confidence:")
print(f"   Rata-rata: {test_data['confidence'].mean():.3f}")
print(f"   Minimum: {test_data['confidence'].min():.3f}")
print(f"   Maksimum: {test_data['confidence'].max():.3f}")
print(f"   Std Dev: {test_data['confidence'].std():.3f}")

# =============================================================================
# 6. SIMPAN HASIL PREDIKSI KE FILE CSV
# =============================================================================
print("\n6. MENYIMPAN HASIL PREDIKSI...")
print("=" * 50)

# Buat DataFrame hasil
results_df = test_data[['id', text_column, 'predicted_label', 'confidence']].copy()
results_df.rename(columns={text_column: 'teks_input'}, inplace=True)

# Simpan ke CSV
output_filename = 'coba_clean.csv'
results_df.to_csv(output_filename, index=False, encoding='utf-8')

print(f"âœ… Hasil prediksi disimpan sebagai: {output_filename}")
print(f"   Total records: {len(results_df)}")
print(f"   Columns: {list(results_df.columns)}")

# =============================================================================
# 7. VISUALISASI DISTRIBUSI PREDIKSI
# =============================================================================
print("\n7. VISUALISASI HASIL...")
print("=" * 50)

plt.figure(figsize=(12, 5))

# Subplot 1: Distribusi Label
plt.subplot(1, 2, 1)
label_counts = results_df['predicted_label'].value_counts().sort_index()
colors = plt.cm.Set3(np.linspace(0, 1, len(label_counts)))
bars = plt.bar(label_counts.index, label_counts.values, color=colors)
plt.title('Distribusi Prediksi Label\n', fontweight='bold')
plt.xlabel('Pilar Astacita')
plt.ylabel('Jumlah Prediksi')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', alpha=0.3)

# Tambahkan nilai di atas bar
for bar, count in zip(bars, label_counts.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1, 
             f'{count}', ha='center', va='bottom')

# Subplot 2: Distribusi Confidence
plt.subplot(1, 2, 2)
plt.hist(results_df['confidence'], bins=20, alpha=0.7, color='skyblue', edgecolor='black')
plt.title('Distribusi Confidence Score\n', fontweight='bold')
plt.xlabel('Confidence Score')
plt.ylabel('Frekuensi')
plt.grid(alpha=0.3)
plt.axvline(results_df['confidence'].mean(), color='red', linestyle='--', 
            label=f'Mean: {results_df["confidence"].mean():.3f}')
plt.legend()

plt.tight_layout()
plt.savefig('prediction_distribution_clean.png', dpi=300, bbox_inches='tight')
plt.show()

print("âœ… Visualisasi disimpan sebagai 'prediction_distribution_clean.png'")

# =============================================================================
# 8. SIMPAN HASIL DALAM FORMAT YANG SIAP DISUBMIT
# =============================================================================
print("\n8. FORMAT HASIL UNTUK SUBMIT...")
print("=" * 50)

# Format sederhana untuk submit (hanya id dan label)
submit_df = results_df[['id', 'predicted_label']].copy()
submit_df.columns = ['id', 'label']  # Rename kolom

submit_filename = 'esults_clean.csv'
submit_df.to_csv(submit_filename, index=False)

print(f"âœ… File submit disimpan sebagai: {submit_filename}")
print(f"   Format: id,label")
print("\nSample submission:")
print(submit_df.head())

print("\n" + "=" * 60)
print("ðŸŽ‰ PREDIKSI SELESAI!")
print("=" * 60)
print(f"Total data diprediksi: {len(results_df)}")
print(f"File hasil: {output_filename}")
print(f"File submit: {submit_filename}")
```

```python
# =============================================================================
# TRAINING ULANG MODEL ASTACITA (TARGET AKURASI > 0.7)
# =============================================================================

import pandas as pd
import numpy as np
import re
import joblib

from sklearn.model_selection import train_test_split, StratifiedKFold, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

# -----------------------------------------------------------------------------
# 1. LOAD DATASET
# -----------------------------------------------------------------------------
print("1. MEMUAT DATASET...")
df = pd.read_csv("coba_clean.csv")   # pastikan ada kolom teks & label
print(df.head())

# Tentukan nama kolom
text_col = [col for col in df.columns if 'teks' in col.lower() or 'text' in col.lower()][0]
label_col = [col for col in df.columns if 'label' in col.lower()][0]

print(f"âœ… Kolom teks: {text_col}, Kolom label: {label_col}")

# -----------------------------------------------------------------------------
# 2. PREPROCESS TEKS (OPSIONAL: STEMMING/STOPWORDS)
# -----------------------------------------------------------------------------
def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', ' ', text)  # buang simbol
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df[text_col] = df[text_col].apply(clean_text)

# -----------------------------------------------------------------------------
# 3. SPLIT DATA
# -----------------------------------------------------------------------------
X = df[text_col]
y = df[label_col]

x_train, x_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# -----------------------------------------------------------------------------
# 4. TF-IDF VECTORIZER (LEBIH KUAT)
# -----------------------------------------------------------------------------
vectorizer = TfidfVectorizer(
    ngram_range=(1,3),     # unigram, bigram, trigram
    max_features=50000,    # lebih banyak fitur
    min_df=3,              # kata minimal muncul di 3 dokumen
    max_df=0.9             # buang kata yang terlalu sering muncul
)
X_train_tfidf = vectorizer.fit_transform(x_train)
X_test_tfidf  = vectorizer.transform(x_test)

print("âœ… TF-IDF selesai. Shape:", X_train_tfidf.shape)

# -----------------------------------------------------------------------------
# 5. COBA BEBERAPA MODEL + GRIDSEARCH
# -----------------------------------------------------------------------------
models = {
    "LogisticRegression": LogisticRegression(max_iter=5000, class_weight="balanced"),
    "LinearSVC": LinearSVC(class_weight="balanced"),
    "RandomForest": RandomForestClassifier(class_weight="balanced", n_jobs=-1, random_state=42),
    "XGBoost": XGBClassifier(
        objective="multi:softmax",
        num_class=len(np.unique(y)),
        eval_metric="mlogloss",
        n_jobs=-1,
        random_state=42
    )
}

# Param grid sederhana (bisa diperbesar kalau mau lebih optimal)
param_grids = {
    "LogisticRegression": {"C": [0.1, 1, 10]},
    "LinearSVC": {"C": [0.1, 1, 10]},
    "RandomForest": {"n_estimators": [100, 300], "max_depth": [None, 20, 50]},
    "XGBoost": {"n_estimators": [200, 500], "max_depth": [6, 10]}
}

best_model = None
best_score = 0
best_name = ""

print("\nðŸ”Ž MULAI TRAINING DAN GRIDSEARCH...")
for name, model in models.items():
    print(f"\n=== {name} ===")
    grid = GridSearchCV(
        model,
        param_grids[name],
        cv=3,
        scoring="accuracy",
        n_jobs=-1,
        verbose=1
    )
    grid.fit(X_train_tfidf, y_train)
    y_pred = grid.predict(X_test_tfidf)
    acc = accuracy_score(y_test, y_pred)

    print(f"Accuracy: {acc:.4f}")
    print("Best Params:", grid.best_params_)
    print("Classification Report:\n", classification_report(y_test, y_pred))

    if acc > best_score:
        best_score = acc
        best_model = grid.best_estimator_
        best_name = name

# -----------------------------------------------------------------------------
# 6. SIMPAN MODEL & VECTORIZER
# -----------------------------------------------------------------------------
print("\nðŸ”¥ MODEL TERBAIK:", best_name)
print(f"ðŸ”¥ Akurasi terbaik di data test: {best_score:.4f}")

joblib.dump(best_model, "best_model_astacita.pkl")
joblib.dump(vectorizer, "best_vectorizer_astacita.pkl")
print("âœ… Model & vectorizer berhasil disimpan!")
```

```python
import pandas as pd
import numpy as np
import re
from collections import Counter

# Preprocessing NLP
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import joblib

# =============================================================================
# 1. LOAD DATA
# =============================================================================
data = pd.read_csv("coba_clean.csv")

text_col = "teks_input"       # sesuaikan dengan file kamu
label_col = "predicted_label" # sesuaikan dengan file kamu

print("Contoh data:")
print(data.head())

# =============================================================================
# 2. CEK DISTRIBUSI LABEL
# =============================================================================
print("\nDistribusi label:")
print(data[label_col].value_counts())

# Kalau distribusi sangat timpang â†’ perlu balancing
# (nanti bisa dicoba oversampling / SMOTE jika perlu)

# =============================================================================
# 3. PREPROCESSING TEKS
# =============================================================================
factory = StemmerFactory()
stemmer = factory.create_stemmer()

stopwords = set(["yang","dan","di","ke","dari","untuk","atau","pada","dengan","sebagai","ini","itu"])

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+|www\S+|https\S+", " ", text) # hapus URL
    text = re.sub(r"[^a-zA-Z\s]", " ", text)             # hapus angka & simbol
    tokens = text.split()
    tokens = [w for w in tokens if w not in stopwords]   # hapus stopwords
    text = " ".join(tokens)
    text = stemmer.stem(text)                            # stemming
    return text

data["clean_text"] = data[text_col].apply(clean_text)

print("\nContoh teks setelah preprocessing:")
print(data["clean_text"].head())

# =============================================================================
# 4. TF-IDF VECTORIZER
# =============================================================================
vectorizer = TfidfVectorizer(
    ngram_range=(1,3), 
    max_features=30000, 
    min_df=2, 
    max_df=0.9
)

X = vectorizer.fit_transform(data["clean_text"])
y = data[label_col]

# =============================================================================
# 5. SPLIT DATA
# =============================================================================
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# =============================================================================
# 6. TRAIN MODEL (Logistic Regression sebagai baseline)
# =============================================================================
model = LogisticRegression(max_iter=2000, class_weight="balanced", C=3)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
acc = accuracy_score(y_test, y_pred)

print("\n=== HASIL EVALUASI ===")
print(f"Accuracy: {acc:.4f}")
print(classification_report(y_test, y_pred))

# =============================================================================
# 7. SIMPAN MODEL + VECTORIZER
# =============================================================================
joblib.dump(model, "improved_model.pkl")
joblib.dump(vectorizer, "improved_vectorizer.pkl")
print("\nâœ… Model dan vectorizer berhasil disimpan!")
```

```python
# =============================================================================
# 1. Import Library
# =============================================================================
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score
import re

# =============================================================================
# 2. Load Data
# =============================================================================
data = pd.read_csv("coba_clean.csv")

# =============================================================================
# 3. Relabeling sederhana dengan keyword
# =============================================================================
def relabel(text):
    text = text.lower()
    if re.search(r"(pancasila|nasionalisme|ideologi|jati diri)", text):
        return "ideologi"
    elif re.search(r"(tentara|militer|perang|pertahanan|keamanan)", text):
        return "pertahanan"
    elif re.search(r"(kerja|pekerjaan|pengangguran|buruh|karyawan|lowongan)", text):
        return "pekerjaan"
    elif re.search(r"(pendidikan|guru|murid|skill|keterampilan|sdm)", text):
        return "sdm"
    elif re.search(r"(hilirisasi|industri|manufaktur|produksi|pengolahan)", text):
        return "hilirisasi"
    elif re.search(r"(pemerataan|kesenjangan|distribusi|keadilan sosial)", text):
        return "pemerataan"
    elif re.search(r"(reformasi|perubahan|demokrasi|transparansi)", text):
        return "reformasi"
    elif re.search(r"(kerukunan|toleransi|damai|harmoni|persatuan)", text):
        return "harmoni"
    else:
        return None

data["final_label"] = data["teks_input"].astype(str).apply(relabel)

# kalau gak ketemu keyword, pakai label lama
data["final_label"] = data["final_label"].fillna(data["predicted_label"])

# =============================================================================
# 4. Balance dataset (5000 total, 625 per label)
# =============================================================================
balanced_data = (
    data.groupby("final_label", group_keys=False)
    .apply(lambda x: x.sample(n=625, replace=True, random_state=42))
)

# =============================================================================
# 5. Train-Test Split + TF-IDF + Model
# =============================================================================
X_train, X_test, y_train, y_test = train_test_split(
    balanced_data['teks_input'], balanced_data['final_label'],
    test_size=0.2, random_state=42, stratify=balanced_data['final_label']
)

vectorizer = TfidfVectorizer(ngram_range=(1,2), max_features=20000)
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)

model = LogisticRegression(max_iter=2000)
model.fit(X_train_tfidf, y_train)

# =============================================================================
# 6. Evaluasi Akurasi
# =============================================================================
y_pred = model.predict(X_test_tfidf)
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
