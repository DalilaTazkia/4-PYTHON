```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('movie_sample_dataset.csv', encoding = 'utf-8')
df.head(50)
```

```python
df.drop(['color', 'language'], axis=1, inplace=True)
df.head()
```

```python
df.isnull()
```

```python
#mengetahui missing vallue dengan atribut
df.isnull().any()
```

```python
#menjumlahkan missing vallue per atribut
df.isnull().sum()
```

```python
#menjumlahkan keseluruhan missing vallue per atribut
df.isnull().sum().sum()
```

```python
#melihat data yang kesalahan dimana 
df['director_name'].unique()
```

```python

df['director_name']=np.where(df['director_name']=='N/A','', df['director_name'])
df['director_name']=np.where(df['director_name']=='Nan','', df['director_name'])
df['director_name']=np.where(df['director_name']=='NaN','', df['director_name'])
df['director_name']=np.where(df['director_name']=='Null','', df['director_name'])
df.head(50)
```
